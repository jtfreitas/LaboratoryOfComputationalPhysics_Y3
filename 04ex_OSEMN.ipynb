4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
CC = pd.read_csv("credit_card.dat", header=None)

def credit_card_reader(number):
    credit_card = "" #set an empty string to append the digits to as strings.
    if len(number) == 118: #only run where the length of the binary string is 118 characters (114 bits + 4 bits padding)
        for n in range(int((len(number)-4)/6)):                #this range specification grants that n will range to 114/6 = 19, which is 16 (6-bit) digits plus 3 spaces. The -4 removes the padding at the end of each line.
            credit_card += chr(int(number[6*n:6*(n+1)],2))     #converts every 6 binary digits into the corresponding Unicode character, appends to the credit card.
    return credit_card

credit_cards = pd.DataFrame(CC[0].apply(credit_card_reader)) #applies the function to the 0-th column of the dataframe.
credit_cards = credit_cards.head(-1) #removes the last line of the dataframe which is just padding.
credit_cards = credit_cards.rename(columns = {0 : "Credit card number"}) #renames the column of the dataframe for fancier formatting
credit_cards.reset_index(drop = True) #resets the index counting for fancier formatting
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
import json

path = "user_data.json"

with open(path) as infile:
    data = json.load(infile) #json.load() opens the file into a list data
    
    #each entry of data is a dictionary, so we keep the format by defining a dictionary that we will populate with lists of the entries as follows:
    
    types = {}
    
    types["ID"] = [i["ID"] for i in data if i["CreditCardType"]=="American Express"]
    types["JobTitle"] = [i["JobTitle"] for i in data if i["CreditCardType"]=="American Express"]
    types["EmailAddress"] = [i["EmailAddress"] for i in data if i["CreditCardType"]=="American Express"]
    types["FirstNameLastName"] = [i["FirstNameLastName"] for i in data if i["CreditCardType"]=="American Express"]
    types["CreditCard"] = [i["CreditCard"] for i in data if i["CreditCardType"]=="American Express"]
    types["CreditCardType"] = [i["CreditCardType"] for i in data if i["CreditCardType"]=="American Express"]
   
    #each types dictionary entry is a list of the elements that exist in data, for which the key CreditCardType is American Express.
    
    filtered_cards = np.array([types["ID"], types["JobTitle"], types["EmailAddress"], types["FirstNameLastName"], types["CreditCard"],types["CreditCardType"]], dtype = str).T
    
    #we create an array with each of the lists in the dictionary type. We transpose it for ease of readability, and to help with the formatting. This method foregoes the dictionary keys, but if that was important we prboably would have just left the file as a .json
    
    np.savetxt("filtered_cards.csv", filtered_cards, delimiter = ",", fmt = '%s')
```

6\. Load the file from this url: [https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1](https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1) with Pandas. 
+ Explore the data (see the info of the data)
+ Draw the istogram of the 'class' field. Decribe what you see


```python
import urllib.request
import matplotlib.pyplot as plt

url = 'https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1'
with urllib.request.urlopen(url) as data_file:
    data_mushrooms = pd.read_csv(data_file)
    
data_mushrooms
```

```python
#Note: I am using the %matplotlib widget magic to allow interactive plots.


ones = data_mushrooms["class"][data_mushrooms["class"] == 1]
zeros = data_mushrooms["class"][data_mushrooms["class"] == 0]

#For the sake of nicer formatting, I am plotting a bar graph instead of a histogram, treating the data as categorical (since we only have two classes).

bar_graph = {"class 0": len(zeros), "class 1": len(ones)}
names = list(bar_graph.keys())
values = list(bar_graph.values())

fig = plt.figure()
plt.bar(names, values, color = ['blue', 'orange'])
plt.title("Class distribution of mushrooms")
fig.show()
```

7\. Load the remote file [https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1](https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1) with Pandas and plot a scatter plot all possible combination of the following fields:
    
  + features_1
  + features_2
  + features_3

```python
import urllib.request
import matplotlib.pyplot as plt

url = 'https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1'

with urllib.request.urlopen(url) as data_file:
    data = pd.read_csv(data_file)
data
```

```python
#Making a figure with six (3! for all the combos of the features) plot elements

fig, axs = plt.subplots(2,3, constrained_layout = True)

fig.suptitle("Plots of $features_1, features_2, features_3$")

features_1 = data["features_1"]
features_2 = data["features_2"]
features_3 = data["features_3"]

axs[0,0].scatter(features_1, features_2, c = features_3)
axs[0,0].set_xlabel("features_1")
axs[0,0].set_ylabel("features_2")
axs[0,0].set_title("coloring using features_3")

axs[0,1].scatter(features_1, features_3, c = features_2)
axs[0,1].set_xlabel("features_1")
axs[0,1].set_ylabel("features_3")
axs[0,1].set_title("coloring using features_2")

axs[0,2].scatter(features_2, features_1, c = features_3)
axs[0,2].set_xlabel("features_2")
axs[0,2].set_ylabel("features_1")
axs[0,2].set_title("coloring using features_3")

axs[1,0].scatter(features_2, features_3, c = features_1)
axs[1,0].set_xlabel("features_2")
axs[1,0].set_ylabel("features_3")
axs[1,0].set_title("coloring using features_1")

axs[1,1].scatter(features_3, features_1, c = features_2)
axs[1,1].set_xlabel("features_3")
axs[1,1].set_ylabel("features_1")
axs[1,1].set_title("coloring using features_2")

axs[1,2].scatter(features_3, features_2, c = features_1)
axs[1,2].set_xlabel("features_3")
axs[1,2].set_ylabel("features_2")
axs[1,2].set_title("coloring using features_1")
plt.show()
```
